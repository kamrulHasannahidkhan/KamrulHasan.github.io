```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/housing.csv")
```

```python
df
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x =df.drop(['median_house_value'], axis=1)
y =df['median_house_value']
```

```python
x_train, x_text, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data = pd.get_dummies(train_data, columns=['ocean_proximity'], prefix=['ocean_proximity'], drop_first=True)
```

```python
train_data.corr()
```

```python
correlation_matrix = train_data.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms']+1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms']+1)
train_data['population'] = np.log(train_data['population']+1)
train_data['households'] = np.log(train_data['households']+1)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(train_data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='longitude', y='latitude', hue='median_house_value', data=train_data)
plt.show()
```

```python
train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(train_data.corr(),annot=True, cmap="YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

```

```python
x_train = train_data.drop(['median_house_value'], axis=1)
y_train = train_data['median_house_value']

x_train_scaled = scaler.fit_transform(x_train)
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
test_data =  x_text.join(y_test)
```

```python
test_data['total_rooms'] = np.log(test_data['total_rooms']+1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms']+1)
test_data['population'] = np.log(test_data['population']+1)
test_data['households'] = np.log(test_data['households']+1)
```

```python
test_data = pd.get_dummies(test_data, columns=['ocean_proximity'], prefix=['ocean_proximity'], drop_first=True)
test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] =test_data['total_rooms'] /test_data['households']
```

```python
test_data
```

```python
x_text = test_data.drop(['median_house_value'], axis=1)
y_test = test_data['median_house_value']
```

```python
x_test_scaled=scaler.transform(x_text)
reg.score(x_test_scaled,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor()
forest.fit(x_train_scaled,y_train)
forest.score(x_test_scaled,y_test)
```

```python
from sklearn.model_selection import  GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    'n_estimators': [30, 50, 100],
    "min_samples_split": [2, 4],
    "max_depth":[None, 4, 8]

}
grid_search = GridSearchCV(forest, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score=True)
grid_search.fit(x_train_scaled, y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(x_test_scaled,y_test)
```
